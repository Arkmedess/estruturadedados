## 1) Estruturas basicas (resumo de 1 linha)

- Lista: mutavel, ordenada, aceita duplicatas.
- Tupla: imutavel, ordenada, aceita duplicatas.
- Set: mutavel, nao ordenado, sem duplicatas.
- Dicionario: chave-valor, chaves unicas, acesso por chave.

Pegadinha comum: set remove duplicatas automaticamente.

```python
lista = [3, 1, 3, 2]
lista.append(4)
lista.sort()
print('lista:', lista)

tupla = (10, 20, 20)
print('tupla:', tupla)

s = {1, 2, 2, 3}
print('set:', s)

d = {'nome': 'Ana', 'idade': 25}
print('dicionario:', d, '| nome =', d['nome'])
```

## 2) Funcoes, parametros, argumentos, *args, **kwargs e lambda

### Termos que confundem (definicao curta)
- Parametro: nome que aparece na definicao da funcao (`def`).
- Argumento: valor real passado na chamada da funcao.
- `*args`: empacota argumentos posicionais extras em uma tupla.
- `**kwargs`: empacota argumentos nomeados extras em um dicionario.
- `lambda`: funcao anonima de uma linha para operacoes simples.

### Regra pratica para prova
- Se pedir quantidade variavel sem nome, pense em `*args`.
- Se pedir quantidade variavel nomeada (chave=valor), pense em `**kwargs`.
- Ordem comum da assinatura: `def f(a, b=0, *args, **kwargs): ...`

```python
# Exemplo 1: funcao com parametros comuns
def soma(a, b):
    return a + b  # devolve o resultado para quem chamou

# Exemplo 2: parametro com valor padrao
def potencia(base, expoente=2):
    return base ** expoente

# Exemplo 3: *args captura varios argumentos posicionais
def soma_tudo(*args):
    # args vira uma tupla: ex. (1, 2, 3, 4)
    return sum(args)

# Exemplo 4: **kwargs captura argumentos nomeados
def exibir_config(**kwargs):
    # kwargs vira um dicionario: {'tema': 'escuro', 'idioma': 'pt-br'}
    for k, v in kwargs.items():
        print(k, '=>', v)

# lambda: funcao anonima curta
dobro = lambda x: x * 2

nums = [1, 2, 3, 4]
print('soma:', soma(2, 3))
print('potencia(5):', potencia(5))
print('potencia(5, 3):', potencia(5, 3))
print('soma_tudo:', soma_tudo(1, 2, 3, 4))
exibir_config(tema='escuro', idioma='pt-br')

# map aplica a funcao em cada item
print('map lambda:', list(map(dobro, nums)))

# filter mantem apenas itens que atendem a condicao
print('filter pares:', list(filter(lambda x: x % 2 == 0, nums)))

# Desempacotamento (tambem pode cair em prova)
valores = (2, 4)
print('soma(*valores):', soma(*valores))

config = {'tema': 'claro', 'idioma': 'en'}
exibir_config(**config)
```

## 3) NumPy (o essencial)

- np.array cria vetor/matriz.
- shape: formato (linhas, colunas).
- ndim: numero de dimensoes.
- size: total de elementos.
- Vetorizacao: array * 2 sem loop explicito.

```python
import numpy as np

v = np.array([1, 2, 3, 4])
m = np.array([[1, 2], [3, 4]])

print('v:', v)
print('m:\n', m)
print('shape m:', m.shape)
print('ndim m:', m.ndim)
print('size m:', m.size)
print('v * 2:', v * 2)
```

## 4) Pandas (o que mais aparece)

- Series: 1 dimensao (como 1 coluna).
- DataFrame: tabela 2D (linhas e colunas).
- Selecao de coluna: df['coluna']
- Filtro: df[df['idade'] > 26]

```python
import pandas as pd

s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
df = pd.DataFrame({
    'nome': ['Ana', 'Beto', 'Carla'],
    'idade': [25, 30, 28]
})

print('Series:\n', s)
print('\nDataFrame:\n', df)
print('\nColuna nome:\n', df['nome'])
print('\nFiltro idade > 26:\n', df[df['idade'] > 26])
```

## 5) Pilha, Fila e Lista Encadeada (teoria curta + clara)

### Pilha (LIFO)
- Last In, First Out: o ultimo que entra e o primeiro que sai.
- Operacoes: `push` (empilha), `pop` (desempilha), `peek` (consulta topo).

### Fila (FIFO)
- First In, First Out: o primeiro que entra e o primeiro que sai.
- Operacoes: `enqueue` (entra no fim), `dequeue` (sai da frente).

### Lista Encadeada (conceito essencial)
- Cada no tem 2 partes: `dado` e `proximo` (ponteiro/referencia).
- A lista comeca pela `cabeca` (head).
- Para chegar ao fim, percorre no a no (acesso sequencial).

### O que pode cair em prova
- Diferenca Pilha x Fila = ordem de remocao (LIFO x FIFO).
- Em lista encadeada, o acesso por indice e O(n), pois precisa percorrer.
- Inserir/remover no inicio e O(1) quando ja temos a cabeca.

```python
# ---------------------------
# PILHA (LIFO)
# ---------------------------
pilha = []
pilha.append(10)  # push: entra no topo
pilha.append(20)
print('topo (peek):', pilha[-1])  # consulta sem remover
print('pop:', pilha.pop())        # remove o topo (ultimo que entrou)

# ---------------------------
# FILA (FIFO)
# ---------------------------
fila = []
fila.append('A')  # enqueue: entra no fim
fila.append('B')
print('dequeue:', fila.pop(0))  # sai da frente (primeiro que entrou)

# ---------------------------
# LISTA ENCADEADA (NO + PROXIMO)
# ---------------------------
class No:
    def __init__(self, dado):
        self.dado = dado
        self.proximo = None  # referencia para o proximo no

class ListaEncadeada:
    def __init__(self):
        self.cabeca = None  # primeiro no da lista

    def adicionar_no_fim(self, dado):
        novo = No(dado)
        if self.cabeca is None:
            # Se a lista esta vazia, novo no vira a cabeca
            self.cabeca = novo
            return

        # Percorre a lista ate o ultimo no
        atual = self.cabeca
        while atual.proximo is not None:
            atual = atual.proximo

        # Encadeia o novo no no final
        atual.proximo = novo

    def exibir(self):
        # Percurso sequencial: cabeca -> ... -> None
        atual = self.cabeca
        while atual is not None:
            print(atual.dado, end=' -> ')
            atual = atual.proximo
        print('None')

le = ListaEncadeada()
le.adicionar_no_fim(1)
le.adicionar_no_fim(2)
le.adicionar_no_fim(3)
le.exibir()
```
